```python
#Number of words present in given file(Wordcount)
count = 0;  
   
#Opens a file in read mode  
file = open("demo_txt", "r")  
      
#Gets each line till end of file is reached  
for line in file:  
    #Splits each line into words  
    words = line.split(" ");  
    #Counts each word  
    count = count + len(words);  
   
print("Number of words present in given file: " + str(count));  
file.close();  
```

```python
#polarity,subjectivity scores
polarity_score=0.131-0.022/0.131+0.022+0.1
subjectivity_score=0.131+0.022/count+0.1
print('The polarity score is',polarity_score)
print('The subjectivity score is',subjectivity_score)
```

```python
#Number of Complex words in the file
def countComplexWords(demo_txt):
    file = open(demo_txt, 'r')
    read_file = file.read().lower()
    words_in_file = read_file.split()  
    count_map = {}
    for i in words_in_file:
        if i in count_map:
            count_map[i] += 1  
        else:
            count_map[i] = 1
    count = 0
    for i in count_map:
        if count_map[i] == 1:
            count += 1
    file.close()
    return count  
print('Number of Complex words in the file are:', 
      countComplexWords('demo_txt'))
  
```

```python
#percentage of complex words
percentage_of_cmplx_word=countComplexWords('demo_txt')/count
print(percentage_of_cmplx_word)
```

```python
#fog index
fog_index=0.4*avg_sentence_length+percentage_of_cmplx_word
print('The fog index is',fog_index)
```

```python
#syllable count
def syllable_count(word):
    count = 0
    vowels = "aeiouy"
    if word[0] in vowels:
        count += 1
    for index in range(1, len(word)):
        if word[index] in vowels and word[index - 1] not in vowels:
            count += 1
            if word.endswith("e"):
                count -= 1
    if count == 0:
        count += 1
    return count
```

```python
print(syllable_count('demo_txt'))
```

```python
#syllable count

def count_syllables(demo_txt):
    word = demo_txt.lower()
    counter = 0
    is_previous_vowel = False
    for index, value in enumerate(word):
        if value in ["a", "e", "i", "o", "u", "y"]:
            if index == len(word) - 1:
                if value == "e":
                    if counter == 0:
                        counter += 1
                else:
                    counter += 1
            else:
                if is_previous_vowel == True:
                    counter += 1
                    is_previous_vowel = False
                    break
            is_previous_vowel = True
        else:
            if is_previous_vowel == True:
                counter += 1
            is_previous_vowel = False
    print ("In \"{}\" I count {} syllables".format(word, counter))

count_syllables("demo_txt")
```
